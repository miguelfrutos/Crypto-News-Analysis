```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_headline(url):

    #grab the information
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    headline = soup.find('h1').text
    url = url
   
    #combine all info into a list of columns
    columns = [headline, url]
    
    #give columns names
    column_names = ['headline', 'url']
    return dict(zip(column_names, columns))
```

#### Inclusion

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

url = "https://finance.yahoo.com/topic/crypto"
included_links =[]
#define a driver
driver = webdriver.Chrome('/Users/isobelimpas/Downloads/chromedriver') #use the path in your computer
driver.maximize_window()
driver.implicitly_wait(5)
#get request
driver.get(url)
driver.find_element(By.XPATH,"//button[@type='submit' and @value='agree']").click() #for cookie consent
for i in range(1,200): 
    try:
    #keep clicking each news article and grab their url
        elem = driver.find_element(By.CSS_SELECTOR, "#Fin-Stream > ul > li:nth-child({}) > div > div > div.Ov\(h\).Pend\(44px\).Pstart\(25px\) > h3 > a".format(i))
    except:
        pass
    href = elem.get_attribute('href')
    driver.execute_script("arguments[0].scrollIntoView();", elem)
    #store in a list
    seen = set(included_links)
    if href not in seen:
        seen.add(href)
        included_links.append(href)
```

```python
included_links
```

```python
info = [get_headline(url) for url in included_links]
included = pd.DataFrame(info)
included['relevance_score'] = 1
included
```
