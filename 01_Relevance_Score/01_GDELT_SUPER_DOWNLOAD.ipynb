## Relevance Score

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://finance.yahoo.com/news/first-mover-asia-defi-resilience-233050642.html"
response = requests.get(url)
soup = BeautifulSoup(response.text)
news = soup.find('div', attrs={'class': 'caas-body'}).text
```

```python
def get_info(url):

    #grab the information
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    news = soup.find('div', attrs={'class': 'caas-body'}).text
    headline = soup.find('h1').text 
    date = datetime.strptime(soup.find('time').text, "%B %d, %Y, %I:%M %p")
    url = url
   
    #combine all info into a list of columns
    columns = [news, headline, date, url]
    
    #give columns names
    column_names = ['news','headline','date', 'url']
    return dict(zip(column_names, columns))
```

#### Inclusion

```python
url = "https://finance.yahoo.com/topic/crypto/"

response = requests.get(url)
soup = BeautifulSoup(response.text,"lxml")

included_links = ['https://finance.yahoo.com'+ soup.select('#Fin-Stream > ul > li:nth-child({}) a'.format(i))[0]['href'] for i in range(1,10)]
```

```python
included_links
```

```python
info = [get_info(url) for url in included_links]
included = pd.DataFrame(info)
included['relevance_score'] = 1
included
```

```python
url = "https://finance.yahoo.com/topic/stock-market-news/"

response = requests.get(url)
soup = BeautifulSoup(response.text,"lxml")

ex_links = ['https://finance.yahoo.com'+ soup.select('#Fin-Stream > ul > li:nth-child({}) a'.format(i))[0]['href'] for i in range(1,10)]
excluded_links.extend(ex_links)
```

```python
info = [get_info(url) for url in excluded_links]
excluded = pd.DataFrame(info)
excluded['relevance_score'] = 0
excluded
```

```python
relevance_df = pd.concat([included, excluded], ignore_index=True, axis=0)
relevance_df
```
